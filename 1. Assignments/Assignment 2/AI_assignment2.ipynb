## Check for Categorical columns and perform encoding.

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

df[['species', 'island', 'sex']] = encoder.fit_transform(df[['species', 'island', 'sex']])
```

```python
df.head()
```

## Split the data into dependent and independent variables.

```python
x = df.iloc[:,1:]
x.head()
```

```python
y = df.iloc[:,0]
y.head()
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler = StandardScaler()

# Scale the independent variables
x = scaler.fit_transform(x)

# Display the scaled data
print("Scaled Data:")
print(x)
```

## Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=34)
```

## Build, Train & Test the model 

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression() # Create an instance of the model
lr.fit(x_train,y_train) # Train the model
lr_y_pred = lr.predict(x_test) # Make predictions
```

```python
from sklearn.svm import SVC
svc = SVC() # Create an instance of the model
svc.fit(x_train,y_train) # Train the model
svc_y_pred = svc.predict(x_test) # Make predictions
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
rf = RFC() # Create an instance of the model
rf.fit(x_train,y_train) # Train the model
rf_y_pred = rf.predict(x_test) # Make predictions
```

## Measure the performance using Evaluation Metrics. 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Logistic Regression
print ("Accuracy Score of Logistic Regression:",accuracy_score(y_test,lr_y_pred))
print ("Confusion matrix of Logistic Regression:",confusion_matrix(y_test,lr_y_pred))
print ("classification_report of Logistic Regression:", classification_report(y_test,lr_y_pred)) 
```

```python
# Support Vector Machine
print ("Accuracy Score of Support Vector Machine:",accuracy_score(y_test,svc_y_pred))
print ("Confusion matrix of Support Vector Machine:",confusion_matrix(y_test,svc_y_pred))
print ("classification_report of Support Vector Machine:", classification_report(y_test,svc_y_pred))
```

```python
# Random Forest Classifier
print ("Accuracy Score of Random Forest Classifier:",accuracy_score(y_test,rf_y_pred))
print ("Confusion matrix of Random Forest Classifier:",confusion_matrix(y_test,rf_y_pred))
print ("classification_report of Random Forest Classifier:", classification_report(y_test,rf_y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Create an instance of the KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(x_train, y_train)

# Make predictions
knn_y_pred = knn.predict(x_test)

print ("Accuracy Score of KNN:",accuracy_score(y_test,knn_y_pred))
print ("Confusion matrix of KNN:",confusion_matrix(y_test,knn_y_pred))
print ("classification_report of KNN:", classification_report(y_test,knn_y_pred))
```
